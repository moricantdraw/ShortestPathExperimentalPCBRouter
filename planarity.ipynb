```python
import numpy as np
import multinetx as mx

# network for each layer
N = 5
g1 = mx.complete_graph(N)
g2 = mx.complete_graph(N)
g3 = mx.complete_graph(N)

# layers interconnection (sparse matrix = no edges)
adj_block = mx.lil_matrix(np.zeros((N * 3, N * 3)))

# define layer interconnection as identity matrices (same node connected to itself across all layers)
adj_block[0:N, N : 2 * N] = np.identity(N)  # L_12
adj_block[0:N, 2 * N : 3 * N] = np.identity(N)  # L_13
adj_block[N : 2 * N, 2 * N : 3 * N] = np.identity(N)  # L_23

adj_block += adj_block.T  # use symmetric inter-adjacency matrix
```
